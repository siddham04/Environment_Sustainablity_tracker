# One Hot Encoding 

```python
import numpy as np
import pandas as pd 
```

```python
df=pd.read_csv('cars.csv')
df
```

```python
df['brand'].unique()
```

# 1) One Hot Encoding using Pandas 

```python
pd.get_dummies(df,columns=['fuel','owner'])
```

# 2) K-1 Encoding

```python
pd.get_dummies(df,columns=['fuel','owner'],drop_first=True)
```

```python
'''we dont use pandas for one hot encoding in  our models because pandas do not remember that which column is placed in which 
position. So, for this we use sklearn library.'''
```

# 3)One Hot Encoding by sklearn library 

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df.iloc[:,0:4],df.iloc[:,-1],test_size=0.2,random_state=2)
```

```python
x_train.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe=OneHotEncoder()
```

```python
ohe.fit(x_train[['fuel','owner']])
```

```python
x_train_new=ohe.transform(x_train[['fuel','owner']]).toarray()
```

```python
# can also do this instead of above two lines
# ohe.fit_transform(x_train[['fuel','owner']]).toarray()
```

```python
x_test_new=ohe.fit_transform(x_test[['fuel','owner']]).toarray()
```

```python
x_test_new
```

```python
np.hstack((x_train[['brand','km_driven']].values,x_train_new))
```

```python
np.hstack((x_train[['brand','km_driven']].values,x_train_new)).shape
```
